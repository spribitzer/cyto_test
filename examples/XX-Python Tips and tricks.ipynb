### Resizing figures

Resizing figure output depends on the plotting package used. See below for specifics for common plotting frameworks (or wrappers like scanpy):

#### matplotlib figure size

For matplotlib and seaborn plots (since seaborn is built on top of mpl), we can use the `.rcParams` method to change plot sizes.

Once set, this applies to all figures plotted afterwards, so it can be set globally at the beginning of a notebook, or included in each code chunk for granular control.

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('mpg')
```

```python
plt.rcParams['figure.figsize'] = [8, 8]
```

```python
plt.scatter(df['weight'], df['mpg'])
```

```python
plt.rcParams['figure.figsize'] = [12, 4]

sns.scatterplot('weight', 'mpg', data = df)
```

#### scanpy figure size

scanpy provides a convenient `set_figure_params()` function to set global parameters related to figure size.

```python
import scanpy as sc
adata = sc.read_h5ad('adata_files/joint_data.h5ad')
```

```python
sc.set_figure_params(figsize=(12, 6))

sc.pl.scatter(adata, x = 'n_genes', y = 'n_umis')
```

### Timing code chunks

```python
%%time

print("This should be quick")
```
