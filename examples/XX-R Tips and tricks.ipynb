### Resizing figures

To resize figure outputs in a Jupyter Notebook, use the `options()` function.

This works for both base plots, and other plotting frameworks like `ggplot2`.

Once set, this applies to all figures plotted afterwards, so it can be set globally at the beginning of a notebook, or included in each code chunk for granular control.

```r
options(repr.plot.width = 8, repr.plot.height = 8)
```

```r
plot(mtcars$wt, mtcars$mpg)
```

```r
options(repr.plot.width = 12, repr.plot.height = 4)

hist(mtcars$wt)
```
